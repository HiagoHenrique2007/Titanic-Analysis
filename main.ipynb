```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
titanic_df = pd.read_csv('./database/titanic.csv')
```

```python
'''
Verificar numero de valores nulos das colunas
'''

columns_df = titanic_df.isna().sum()
labels = columns_df.index.to_list()
values = columns_df.to_numpy().tolist()

plt.figure(figsize=(16, 6))
plt.bar(labels, values)
```

```python
'''
Varificar o a distribuição de idades
'''

ages_series = titanic_df['Age']

plt.figure(figsize=(16, 6))
plt.hist(ages_series.to_numpy().tolist())
plt.show()


```

```python
'''
Exibir os sobreviventes por classe e sexo
'''

number_survivors_male = titanic_df['Sex'][ titanic_df['Sex'] == 'male' ].shape[0]

number_survivors_female = titanic_df['Sex'][ titanic_df['Sex'] == 'female' ].shape[0]
number_survivors_male, number_survivors_female

number_survivors_class = [ titanic_df['Pclass'][ titanic_df['Pclass'] == class_number ].shape[0] for class_number in range(1,4)]

labels_of_class = titanic_df['Pclass'].drop_duplicates().sort_values().to_list()

for index, label_class in enumerate(labels_of_class):
  labels_of_class[index] = f'Class {label_class}'

fig, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[0].bar(['Male', 'Female'], [number_survivors_male, number_survivors_female], label='Graphc per Sex')
axes[0].set_title('Survivors per Sex')
axes[0].set_xlabel('Sex')
axes[0].set_ylabel('Howmatch')
axes[0].legend()

axes[1].bar(labels_of_class, number_survivors_class, color='red', label='Graphc per Class')
axes[1].set_title('Survivors per Class')
axes[1].set_xlabel('Class')
axes[1].set_ylabel('Howmatch')
axes[1].legend()

plt.tight_layout()
plt.show()
```
