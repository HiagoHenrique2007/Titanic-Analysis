```python
import pandas as pd
```

```python
titanic_df = pd.read_csv('./database/titanic.csv')
```

```python
columns = titanic_df.columns
columns_with_nulls = {}
for column in columns:
  column_values = titanic_df[column]
  null_values = column_values[ column_values.isna() ]
  number_of_nulls = null_values.shape
  
  if number_of_nulls[0] != 0:
    columns_with_nulls.update({ f'{column}': number_of_nulls[0] })
    print(f'{column} has null values')



columns_with_nulls
  
```

```python
'''
Verificar numero de valores nulos das colunas
'''

import matplotlib.pyplot as plt


columns_df = titanic_df.isna().sum()
labels = columns_df.index.to_list()
values = columns_df.to_numpy().tolist()

plt.figure(figsize=(16, 6))
plt.bar(labels, values)
plt.savefig('./reports/columns_with_nulls.png')
```

```python
'''
Varificar o a distribuição de idades
'''

ages_series = titanic_df['Age']

plt.figure(figsize=(16, 6))
plt.hist(ages_series.to_numpy().tolist())
plt.savefig('./reports/ages.png', dpi=300)
plt.show()


```
